```python
## Ecrire la fonction « extrait_comp.py » qui extrait les trois composantes C1, C2, C3 d’une image et les afficher séparément. La fonction doit générer finalement une
## figure composée de quatre images (ex : pour le RGB : l’originale, l’image en rouge,
## l’image en vert et l’image en bleue)

def extrait_comp(clown):
  # extrait des 3 composantes 
  
  c1 = clown[:,:,0]
  c2 = clown[:,:,1]
  c3 = clown[:,:,2]

  # Afficher c1,c2,c3

  figure = plt.figure(figsize=(7,10))
  
  figure.add_subplot(3,3,1)
  plt.imshow(c1)
  plt.axis('off')
  plt.title("C1")
  
  figure.add_subplot(3,3,2)
  plt.imshow(c2)
  plt.axis('off')
  plt.title("C2")

  figure.add_subplot(3,3,3)
  plt.imshow(c3)
  plt.axis('off')
  plt.title("C3")

extrait_comp(clown)  




```

# Transformation spectrale : Transformation de Fourier

# a - Transformation de fourier

```python
## Import Images
crr1 = plt.imread("/content/drive/MyDrive/images2021/carre1.png")
crr2= plt.imread("/content/drive/MyDrive/images2021/carre2.png")
```

```python
f = np.fft.fft2(crr1)
sf = np.abs(f)
sfcrr1 = np.fft.fftshift(sf)
sfcrr2 = np.fft.fftshift(np.abs(np.fft.fft2(crr2)))
```

```python
plt.figure()
plt.subplot(221),plt.imshow(crr1, cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(crr2, cmap = 'gray')
plt.xticks([]), plt.yticks([])

plt.subplot(223),plt.imshow(sfcrr1, cmap='jet')
plt.title('Magnitude Spectrum carre1'), plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(sfcrr2, cmap='jet')
plt.title('Magnitude Spectrum carre2'), plt.xticks([]), plt.yticks([])

fig=plt.figure()
xx1, yy1 = np.mgrid[0:sfcrr1.shape[0], 0:sfcrr1.shape[1]]
ax = fig.gca(projection='3d')
ax.plot_surface(xx1, yy1, sfcrr1, cmap='jet')

fig = plt.figure()
xx2, yy2 = np.mgrid[0:sfcrr2.shape[0], 0:sfcrr2.shape[1]]
ax = fig.gca(projection='3d')
ax.plot_surface(xx2, yy2, sfcrr2, cmap='jet')
```
