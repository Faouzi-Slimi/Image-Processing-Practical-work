```python
# Import library
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
# lire l'image de niveau gray lena.pmb
im1=plt.imread("/content/drive/MyDrive/images2021/lena.bmp")
```

```python
# dimention
m,n = im1.shape
```

```python
# L'image negative
imn = 255 - im1
```

```python
# SYMETRYE 
imv = im1[:,::-1]
imh = im1[::-1,:]
imc = im1[::-1,::-1]
```

# b- Operation Logique

```python
# Lire Imges : cercle,rectangle et Triangle

cr = plt.imread("/content/drive/MyDrive/images2021/cercle.bmp")
cr.shape
cr = cr[:,:,0]/255

rec = plt.imread("/content/drive/MyDrive/images2021/rectangle.bmp")
rec = rec[:,:,0]/255
tri = plt.imread("/content/drive/MyDrive/images2021/triangle.bmp")
tri = tri[:,:,0]/255
```

```python
#Visualisation
plt.close()
plt.figure()
plt.subplot(231)
plt.imshow(cr,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(232)
plt.imshow(rec,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(233)
plt.imshow(tri,cmap='gray')
plt.xticks([]), plt.yticks([])
```

```python
## OU AND NOT
im1 = np.logical_and(np.logical_and(cr,rec),tri)*1
# ou bien d=d.asype(int)
im2 = np.logical_or(np.logical_or(cr,rec),tri)*1
im3 = np.logical_not(np.logical_not(cr,rec),tri)*1
```

```python
# Visualisation
plt.subplot(234)
plt.imshow(im1,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(235)
plt.imshow(im2,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(236)
plt.imshow(im3,cmap='gray')
plt.xticks([]), plt.yticks([])
```
